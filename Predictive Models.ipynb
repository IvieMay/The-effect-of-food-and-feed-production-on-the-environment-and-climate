```python
# Summary statistics for both datasets
summary_fp = df_fp.describe()
summary_fao = df_fao.describe()

# Select only numeric columns for correlation calculation
numeric_cols_fp = df_fp.select_dtypes(include='number').columns
numeric_cols_fao = df_fao.select_dtypes(include='number').columns

# Correlation matrix for both datasets
correlation_fp = df_fp[numeric_cols_fp].corr()
correlation_fao = df_fao[numeric_cols_fao].corr()
```

```python
from sklearn.model_selection import train_test_split
from sklearn.linear_model import LinearRegression
from sklearn.ensemble import RandomForestRegressor, GradientBoostingRegressor
from sklearn.svm import SVR
from sklearn.preprocessing import StandardScaler
from sklearn.metrics import mean_squared_error, mean_absolute_error, r2_score
import matplotlib.pyplot as plt
import seaborn as sns

# Prepare the dataset for modeling
features = df_fp[['Land use change', 'Animal Feed', 'Farm', 'Processing', 'Transport', 'Packging', 'Retail']]
target = df_fp['Total_emissions']

# Split the data into training and testing sets
X_train, X_test, y_train, y_test = train_test_split(features, target, test_size=0.3, random_state=42)

# Standardize the features for models that require it
scaler = StandardScaler()
X_train_scaled = scaler.fit_transform(X_train)
X_test_scaled = scaler.transform(X_test)

# Initialize models
models = {
    "Random Forest": RandomForestRegressor(n_estimators=100, random_state=42),
    "Gradient Boosting": GradientBoostingRegressor(n_estimators=100, random_state=42),
    "Support Vector Regression": SVR(kernel='rbf', C=1.0, epsilon=0.1)
}

# Train and evaluate models
results = {}
for name, model in models.items():
    if name in ["Support Vector Regression"]:
        model.fit(X_train_scaled, y_train)
        y_pred = model.predict(X_test_scaled)
    else:
        model.fit(X_train, y_train)
        y_pred = model.predict(X_test)
    mse = mean_squared_error(y_test, y_pred)
    mae = mean_absolute_error(y_test, y_pred)
    r2 = r2_score(y_test, y_pred)
    results[name] = {"MSE": mse, "MAE": mae, "R²": r2}

# Print the results line by line
for model_name, metrics in results.items():
    print(f"Model: {model_name}")
    for metric_name, value in metrics.items():
        print(f"  {metric_name}: {value:.4f}")
    print()
```

```python
# Prepare data for plotting
metrics_df = pd.DataFrame(results).T.reset_index().rename(columns={'index': 'Model'})
metrics_melted = metrics_df.melt(id_vars='Model', var_name='Metric', value_name='Value')

# Plot the metrics
plt.figure(figsize=(12, 8))
sns.barplot(x='Metric', y='Value', hue='Model', data=metrics_melted)
plt.title('Model Performance Metrics Comparison')
plt.xlabel('Metric')
plt.ylabel('Value')
plt.legend(title='Model')
plt.show()

# Scatter plot for each model
fig, axes = plt.subplots(2, 2, figsize=(14, 12))
axes = axes.flatten()

for i, (name, model) in enumerate(models.items()):
    if name == "Support Vector Regression":
        y_pred = model.predict(X_test_scaled)
    else:
        y_pred = model.predict(X_test)
    
    axes[i].scatter(y_test, y_pred, alpha=0.3)
    axes[i].plot([y_test.min(), y_test.max()], [y_test.min(), y_test.max()], 'r--', lw=2)
    axes[i].set_title(f"{name} (R²: {results[name]['R²']:.2f})")
    axes[i].set_xlabel('Actual')
    axes[i].set_ylabel('Predicted')

plt.tight_layout()
plt.show()
```
